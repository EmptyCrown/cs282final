```python
import numpy as np
import numpy.random as npr 
import matplotlib.pyplot as plt 
import pdb 
import random

import math
from scipy.stats import poisson
import time

alpha = 5.

def LOF(arr):
    arr = arr[:, np.lexsort(np.flipud(arr))][:,::-1]
    return arr

def indian_buffet(N, alpha):
    width = np.random.poisson(alpha)
    Z = np.zeros((N, width))
    Z[0,:] = np.ones(width)
    for i in range(1, N):
        for j in range(0, width):
            r = random.randint(0, i) # simulate a probability of m_k/i
            if r < sum(Z[:i,j]):
                Z[i,j] = 1
        newWidth = width + np.random.poisson(float(alpha)/(i+1))
        newZ = np.zeros((N, newWidth))
        newZ[:,:width] = Z
        newZ[i,width:] = np.ones(newWidth - width)
        
        Z = newZ
        width = newWidth
    
    num_features = []
    for i in range(0, N):
        num_features.append(sum(Z[i,:]))
    return (LOF(Z), num_features)

# The little blocks from G&G, too easy for most applications but
# useful for debugging algorithms
def generate_gg_blocks(): 
    A = np.zeros( [ 4 , 36 ] ) 
    A[0, [ 1 , 6 , 7 , 8 , 13 ] ] = 1
    A[1, [ 3 , 4 , 5 , 9 , 11 , 15 , 16 , 17  ] ] = 1 
    A[2, [ 18 , 24 , 25 , 30 , 31 , 32 ] ] = 1
    A[3, [ 21 , 22 , 23 , 28 , 34 ] ] = 1 
    return A 
    
# Alternative: Random A.  Note that one should probably do something
# in between that's more structured but less toy for exploring!
def generate_random_A( feature_count , dim_count , sigma_a=10 ):
    A = sigma_a * npr.normal( 0 , 1 , [ feature_count , dim_count ] ) 
    return A 

# Generate the data
def generate_data( data_count , data_type='infinite-random' , sigma=0.1 , sigma_a=5 ):
    # data_type can be 'gg' for the basic blocks, 'finite-random' for
    # a random A but a finite Z (same size as gg), and
    # 'infinite-random' for a simulated Z from the IBP and a random A
    dim_count = 36
    if data_type == 'gg': 
        A = generate_gg_blocks()
        feature_count = A.shape[0]
        dim_count = A.shape[1]
        Z = npr.binomial( 1 , .25 , [ data_count , feature_count ] )
    if data_type == 'finite-random':
        feature_count = 10
        A = generate_random_A( feature_count , dim_count, sigma_a )
        Z = npr.binomial( 1 , .25 , [ data_count , feature_count ] )
    if data_type == 'infinite-random':
        Z , _ = indian_buffet( data_count , alpha)
        feature_count = Z.shape[1]
        A = generate_random_A( feature_count , dim_count, sigma_a ) 
        
    # add noise to the data and return 
    data_set = np.dot( Z , A ) + sigma * npr.normal( 0 , 1 , [ data_count , dim_count ] ) 
    return data_set , Z , A 
```

```python
sigma_x = 0.1
sigma_a = 5.
X, oldZ, A = generate_data(20, 'infinite-random', sigma_x, sigma_a)
I2 = np.identity(oldZ.shape[0])
D = X.shape[1]
```

```python
def add_col_to_Z(Z, n, q):
    newWidth = Z.shape[1] + q
    newZ = np.zeros((Z.shape[0], newWidth))
    newZ[:,:Z.shape[1]] = Z
    newZ[n,Z.shape[1]:] = np.ones(q)
    return newZ

def remove_empty_col(Z):
    newZ = np.copy(Z)
    empty = []
    for k in range(Z.shape[1]):
        if sum(Z[:,k]) == 0:
            empty.append(k)
    newZ = np.delete(newZ, empty, 1)
    return newZ
```

```python
cutoff = 10
    
def collapsed_log_X_given_Z(Z):
    I = np.identity(Z.shape[1])
    N, K = Z.shape
    H1 = np.dot(Z.T, Z) + (sigma_x / sigma_a) ** 2 * I
    H2 = np.dot(np.dot(Z, np.linalg.inv(H1)), Z.T)
    numer = -1. / (2 * sigma_x ** 2) * np.trace(np.dot(np.dot(X.T, I2 - H2), X))
    denom = (N * D / 2) * np.log(2 * np.pi) + ((N - K) * D) * np.log(sigma_x) + (D * K) * np.log(sigma_a) + (D/2) * np.log(np.linalg.det(H1))
    return numer - denom
    
def collapsed_existing_log_prob(Z, n, k, value=1):
    N, K = Z.shape
    Z[n][k] = value
    col = [int(q == value) for q in Z[:, k]]
    p = np.log(float(sum(col) - 1) / N)
    return p + collapsed_log_X_given_Z(Z)

def collapsed_get_newk(inZ, n):
    lp = []
    
    for q in range(0,cutoff):
        Z = add_col_to_Z(inZ, n, q)
        N, K = Z.shape
        log_prob = np.log(poisson.pmf(q, alpha/N)) + collapsed_log_X_given_Z(Z)
        #print log_prob
        lp.append(log_prob)
    lp = np.array(lp)
    lp = lp - max(lp)
    lp = np.exp(lp)
    lp = lp / sum(lp)
    return np.random.choice(np.arange(cutoff), p=lp)
    

def collapsed_update_row(inZ, n):
    Z1 = np.copy(inZ)
    Z2 = np.copy(inZ)
    N, K = Z1.shape
    for k in range(K):
        lp1 = collapsed_existing_log_prob(Z1,n,k,1)
        lp2 = collapsed_existing_log_prob(Z1,n,k,0)
        mlp = max(lp1, lp2)
        p1 = np.exp(lp1 - mlp) / (np.exp(lp1 - mlp) + np.exp(lp2-mlp))
        p2 = 1 - p1
        Z1[n][k] = np.random.binomial(1,p1)
    newk = collapsed_get_newk(Z2, n)
    newZ = add_col_to_Z(Z1, n, newk)
    return newZ

#visualize(add_col_to_Z(oldZ, 10, 20))
visualize(oldZ)
```

```python
def sample_A(Z):
    newA = np.zeros((Z.shape[1], D))
    I = np.identity(Z.shape[1])
    mean = np.dot(np.linalg.inv(np.dot(Z.T, Z) + (sigma_x / sigma_a) ** 2 * I), np.dot(Z.T, X))
    cov = sigma_x ** 2 * np.linalg.inv(np.dot(Z.T, Z) + (sigma_x / sigma_a) ** 2 * I)
    for d in range(D):
        newA[:,d] = np.random.multivariate_normal(mean[:,d], cov)
    
    return newA

def uncollapsed_log_X_given_ZA(Z,A):
    N,K = Z.shape
    I = np.identity(K)
    H1 = X - np.dot(Z,A)
    numer = -1. / (2 * sigma_x ** 2) * np.trace(np.dot(H1.T, H1))
    denom = (N * D / 2) * np.log(2 * np.pi * sigma_x **2) 
    return numer - denom

def uncollapsed_existing_log_prob(Z,n,k,value=1):
    Z[n][k] = value
    A = sample_A(Z)
    N,K = Z.shape
    col = [int(q == value) for q in Z[:, k]]
    p = np.log(float(sum(col) - 1) / N)
    
    return p + uncollapsed_log_X_given_ZA(Z,A)

def uncollapsed_get_newk(inZ, n):
    lp = []
    for q in range(0,cutoff):
        Z = add_col_to_Z(inZ, n, q)
        N,K = Z.shape
        A = sample_A(Z)
        prob = np.log(poisson.pmf(q, alpha/N)) + uncollapsed_log_X_given_ZA(Z, A)
        lp.append(prob)
    lp = np.array(lp)
    lp = lp - max(lp)
    lp = np.exp(lp)
    lp = lp / sum(lp)
    return np.random.choice(np.arange(cutoff), p=lp)

def uncollapsed_update_row(inZ, n):
    Z1 = np.copy(inZ)
    Z2 = np.copy(inZ)
    N,K = Z1.shape
    for k in range(K):
        lp1 = uncollapsed_existing_log_prob(Z1,n,k,1)
        lp2 = uncollapsed_existing_log_prob(Z1,n,k,0)
        mlp = max(lp1, lp2)
        p1 = np.exp(lp1 - mlp) / (np.exp(lp1 - mlp) + np.exp(lp2-mlp))
        p2 = 1 - p1
        Z1[n][k] = np.random.binomial(1,p1)
    newk = uncollapsed_get_newk(Z2, n)
    newZ = add_col_to_Z(Z1, n, newk)
    return newZ

visualize(oldZ)
```
